```python
import os
from pathlib import Path
import pickle
# get constrains from path
PROJECT_ROOT = Path(os.getcwd()).parents[1].resolve()
DATA_PATH = os.path.join(PROJECT_ROOT, 'data/interim')
CONSTRAINT_FILE = 'semantic_sap_sam_filtered_bpmn_declare_const.pkl'

with open(os.path.join(DATA_PATH, CONSTRAINT_FILE), 'rb') as f:
    constraints = pickle.load(f)
constraints = constraints[constraints['Level']=='Activity']
constraints
```

```python
import pm4py
LOG_PATH = os.path.join(PROJECT_ROOT, "data/logs")
LOG_FILE = "BPI_Challenge_2019-3-w-after.xes"
log = pm4py.read_xes(str(os.path.join(LOG_PATH, LOG_FILE)))
```

```python
# compute the duration of all traces and add an attribute to the log: group by case:concept:name and per group compute the difference between the max and min timestamp and convert to minutes

log['duration'] = log.groupby('case:concept:name')['time:timestamp'].transform('max') - log.groupby('case:concept:name')['time:timestamp'].transform('min')
log['duration'] = log['duration'].dt.total_seconds() / 60
log
```

```python
from semconstmining.selection.instantiation.recommendation_config import RecommendationConfig
import pandas as pd
from os.path import exists
from semconstmining.config import Config
from semconstmining.main import get_resource_handler, get_or_mine_constraints, get_parts_of_constraints, \
    get_log_and_info, compute_relevance_for_log, recommend_constraints_for_log
from semconstmining.parsing.label_parser.nlp_helper import NlpHelper

CURRENT_LOG_FILE = "BPI_Challenge_2019-3-w-after.xes"
MODEL_COLLECTION = "semantic_sap_sam_filtered"
config = Config(PROJECT_ROOT, MODEL_COLLECTION)
rec_config = RecommendationConfig(config, semantic_weight=0.9, top_k=250)

nlp_helper = NlpHelper(config)
resource_handler = get_resource_handler(config, nlp_helper)
all_constraints = get_or_mine_constraints(config, resource_handler, min_support=1)
constraints = all_constraints[all_constraints['Level']=='Activity']
nlp_helper.pre_compute_embeddings(sentences=get_parts_of_constraints(config, constraints))
event_log, log_info = get_log_and_info(config, nlp_helper, CURRENT_LOG_FILE)
# Log-specific constraint recommendation
if not exists(config.DATA_INTERIM / (CURRENT_LOG_FILE + "-constraints_with_relevance.pkl")):
    constraints = compute_relevance_for_log(config, constraints, nlp_helper, CURRENT_LOG_FILE, pd_log=event_log, precompute=True)
    constraints.to_pickle(config.DATA_INTERIM / (CURRENT_LOG_FILE + "-constraints_with_relevance.pkl"))
else:
    constraints = pd.read_pickle(
        config.DATA_INTERIM / (CURRENT_LOG_FILE + "-constraints_with_relevance.pkl"))
recommended_constraints = recommend_constraints_for_log(config, rec_config, constraints, nlp_helper, CURRENT_LOG_FILE, pd_log=event_log)
recommended_constraints
```

```python
from semconstmining.selection.consistency.consistency import ConsistencyChecker

consistency_checker = ConsistencyChecker(config)
# Check for trivial inconsistencies
recommended_constraints = consistency_checker.check_trivial_consistency(recommended_constraints)
recommended_constraints
```

```python
from semconstmining.main import check_constraints, get_violation_to_cases, add_original_labels

recommended_constraints = add_original_labels(config, recommended_constraints, log_info)
violations = check_constraints(config, CURRENT_LOG_FILE, recommended_constraints, nlp_helper, pd_log=event_log, with_id=True)
violations_to_cases = get_violation_to_cases(config, violations, with_id=True)
violation_df = pd.DataFrame.from_records(
    [{"violation": violation, "num_violations": len(cases), "cases": cases} for violation, cases in
     violations_to_cases.items()])
merged_df = pd.merge(recommended_constraints.reset_index(), violation_df,
                     left_on=config.RECORD_ID, right_on='violation', how='inner')
merged_df
```

```python
# filter for violations that have at least 100 cases
filtered_df = merged_df[merged_df['num_violations'] < 100]
# obs_id the index
filtered_df = filtered_df.set_index(config.RECORD_ID)
# merge the lists in log_label_left and log_label_right into a column called original_params
filtered_df['original_params'] = filtered_df.apply(lambda x: [*x['log_label_left'], *x['log_label_right']], axis=1)
# only keep rows where there is at least one original param
filtered_df = filtered_df[filtered_df['original_params'].apply(lambda x: len(x) > 0)]
filtered_df
```

```python
# per violation compute the performance of cases that violate if and compare to the performance of cases that do not violate the constraint
def compute_performance_of_cases_with_violation(violation, violations_to_cases, log):
    cases = violations_to_cases[violation]
    # only keep the first row per group when grouped by case:concept:name
    performance = log[log['case:concept:name'].isin(cases)].groupby('case:concept:name').first().reset_index()['duration']
    # transfrom into list
    performance = performance.tolist()
    return performance

def compute_performance_of_cases_without_violation(violation, violations_to_cases, log, activated_cases):
    cases_without_violation = list(set(activated_cases) - set(violations_to_cases[violation]))
    # only keep the first row per group when grouped by case:concept:name
    performance = log[log['case:concept:name'].isin(cases_without_violation)].groupby('case:concept:name').first().reset_index()['duration']
    # transfrom into list
    performance = performance.tolist()
    return performance


def compute_cases_that_activate_constraint(constraint_string, original_params, log):
    # group by cases and get unique activity names (concept:name)
    cases = log.groupby('case:concept:name')['concept:name'].unique()
    # get case ids (case:concept:name) where the constraint is activated, i.e., where one of the original_params is in the unique activity names
    cases = cases[cases.apply(lambda x: original_params[1] in x if len(original_params[1])> 1 and  "Precedence" in constraint_string else original_params[0] in x if len(original_params)>0 else False)].index
    return cases

```

```python
# compute the average performance of cases with and without violation and save it in a dataframe
performance_records = []
for violation in filtered_df['violation']:
    activated_cases = compute_cases_that_activate_constraint(filtered_df.loc[violation, 'constraint_string'], filtered_df.loc[violation, 'original_params'], log)
    print(f"Activated cases: {len(activated_cases)}")
    performance_with_violation = compute_performance_of_cases_with_violation(violation, violations_to_cases, log)
    if len(violations_to_cases[violation])<=len(activated_cases):
        performance_without_violation = performance_with_violation
    else:
        print(f"Violation {violation} is not violated whenever it is activated")
        performance_without_violation = compute_performance_of_cases_without_violation(violation, violations_to_cases, log, activated_cases)
    if len(performance_with_violation) == 0 or len(performance_without_violation) == 0:
        print(f"Violation {violation} has no cases with or without violation")
        continue
    performance_records.append({"violation": violation, "performance_with_violation": sum(performance_with_violation)/len(performance_with_violation), 
                                "performance_without_violation": sum(performance_without_violation)/len(performance_without_violation)})
performance_df = pd.DataFrame.from_records(performance_records)
performance_df


    
```

```python
performance_df['performance_difference'] = performance_df['performance_with_violation'] - performance_df['performance_without_violation']
performance_df
```

```python
merged_performance_df = pd.merge(recommended_constraints.reset_index(), performance_df,
                     left_on=config.RECORD_ID, right_on='violation', how='inner')
```
